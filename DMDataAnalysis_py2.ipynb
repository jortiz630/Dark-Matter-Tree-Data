```python
z = (1/a)-1 # Redshift based on scale factor.
omegaB = 0.049 # Mass fraction in baryons.
omegaM = 0.32 # Mass fraction of all matter.
fStar = 0.05 #fraction of gas used to create stars.
eta = 4000 # Number of ionizing photons produced per baryon.
mProton = 1.6726219*10e-27 # Mass of a proton.
ageUniverse = ((0.93e+09)*((1+z)/7)**(-1.5))*(3.154e+07) # Age of the universe at given redshift in seconds.
```

```python
halosPractice = halosAtScale0_2[:6000]
halosPractice = halosPractice[['x(17)','y(18)','z(19)','mvir(10)']].values
halosPractice[:,:3] *= 3.086e+24
halosPractice[:,3] = halosPractice[:,3]*0.67*mSun
print halosPractice
print halosPractice.dtype
```

```python
def radiusSquared(a,b,c,x,y,z):
    return ((x-a)**2) + ((y-b)**2) + ((z-c)**2)

def SFR(mHalo): # Function that returns star formation rate as a function of its mass.
    return mHalo*(omegaB/omegaM)*(fStar/ageUniverse)

def Lum(mHalo):
    return SFR(mHalo)*(eta/mProton)

def Flux(a,b,c,x,y,z,mHalo):
    return Lum(mHalo)/(4*np.pi*radiusSquared(a,b,c,x,y,z))
```

# Need to run full data set to be able to compute the next few lines of code for the final histogram.

```python
mSun = 1.989e+30
lowerBound = (2e+08)
upperBound = (5e+08)
print lowerBound, upperBound
```

```python
fluxResults = []

for i, j, k, mass in halosPractice:
    fluxSum = 0
    for x, y, z, massValue in halosPractice:
        # print i,j,k,x,y,z, massValue, radiusSquared(i,j,k,x,y,z)
        if (radiusSquared(i,j,k,x,y,z) == 0):
            continue
        else:
            fluxTemp = Flux(i,j,k,x,y,z,massValue)
            # print 'fluxTemp:', fluxTemp
            fluxSum += fluxTemp
    # print fluxSum
    fluxResults.append(fluxSum)
fluxResults = np.array(fluxResults)
print fluxResults
print fluxResults.shape
```

```python
flux0 = 6.7e+06
fluxResultsNew = fluxResults*(1/flux0)
print fluxResultsNew
```

```python
fluxLogs = pd.DataFrame(data = np.log10(fluxResultsNew))
# fluxLogsPlot = fluxLogs[(fluxLogs[0]>3.978e+38) & (fluxLogs[0]<9.945e+38)]
print fluxLogs
```

```python
halosForHist = halosAtScale0_2[['x(17)','y(18)','z(19)','mvir(10)']]
halosForHist = halosForHist.assign(fluxLogs = fluxLogs.values)
print halosForHist
```
