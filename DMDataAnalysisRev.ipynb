```python
redshift = (1/scaleFactor)-1 # Redshift based on scale factor.
omegaB = 0.049 # Mass fraction in baryons.
omegaM = 0.32 # Mass fraction of all matter.
fStar = 0.05 # fraction of gas used to create stars.
eta = 4000 # Number of ionizing photons produced per baryon.
mProton = 1.6726219*10e-27 # Mass of a proton in kilograms.
ageUniverse = ((0.93e+09)*((1+redshift)/7)**(-1.5))*(3.154e+07) # Age of the universe at given redshift in seconds.
ageUniverseYears = ((0.93e+09)*((1+redshift)/7)**(-1.5))
print redshift, omegaB, omegaM, fStar, eta, mProton, ageUniverse, ageUniverseYears
```

```python
mSun = 1.989e+30 # Mass of sun in kilograms.
lowerBound = (2e+08) # Lower bound for halo mass in data range for histogram (kilograms).
upperBound = (5e+08) # Upper bound for halo mass in data range for histogram (kilograms).
minMass = 1.5e+08*((1+redshift)/11)**-1.5 # Minimum mass necessary for star formation to occur (solar masses).
print lowerBound, upperBound, minMass
```

```python
def deltaXsquared(a,x):
    return (x-a)**2
def deltaYsquared(b,y):
    return (y-b)**2
def deltaZsquared(c,z):
    return (z-c)**2
```

```python
def radiusSquared(a,b,c,x,y,z):
    return deltaXsquared(a,x) + deltaYsquared(b,y) + deltaZsquared(c,z)

def SFR(mHalo): # Function that returns star formation rate as a function of its mass.
    return mHalo*(omegaB/omegaM)*(fStar/ageUniverse)

def Lum(mHalo):
    return SFR(mHalo)*(eta/mProton)*0.1

def Flux(a,b,c,x,y,z,mHalo):
    return Lum(mHalo)/(4*np.pi*radiusSquared(a,b,c,x,y,z))
```

```python
halosPractice = halosScale0dot2[['x(17)','y(18)','z(19)','mvir(10)']]
halosPractice = halosPractice[halosPractice['mvir(10)'] > minMass].values
print halosPractice # Halos that will, for sure, have formed stars at this redshift.
```

```python
mass10 = halosScale0dot2.iloc[:,10:11].values
print mass10 # Prints mass of halos at current redshift in solar masses.

sideBox = 20*(3.086e+24)
radiusBox = sideBox/2
```

```python
def SFRannual(mHalo): # Function that returns star formation rate as a function of its mass.
    return mHalo*(omegaB/omegaM)*(fStar/ageUniverseYears)

meanSFR, meanLum, meanSFRannual = 0, 0, 0
for i in mass10:
    meanSFR += SFR(i)
    meanSFRannual += SFRannual(i)
    meanLum += Lum(i)
    
meanSFRannual = meanSFRannual/(20**3)/0.67 # Average solar mass produced annually per Mpc^3.
meanSFR = (meanSFR*mSun)/(sideBox**3)/0.67 # Average kilograms produced per second per cm^3.
meanLum = meanLum/(sideBox**3) # Average number of photons produced per second per cm^3.

print meanSFRannual, np.log10(meanSFRannual), '\n', meanSFR, np.log10(meanSFR), '\n', meanLum
```

```python
externalDistance = (20)*3.086e+24
externalFlux = (meanLum)*(externalDistance - radiusBox) # Calculated flux from outside a distance of 10 Mpc
print externalFlux                                      # from the halo receiving the flux.
```
